```python
import pandas as pd
import time
import json
import requests
from tqdm import tqdm_notebook

api_url='http://ddragon.leagueoflegends.com/cdn/10.6.1/data/en_US/champion.json'
r=requests.get(api_url)
champ=r.json()
```

```python
champ_df=pd.DataFrame(champ['data']).T
champ_df.reset_index(inplace=True)
champ_df.drop('index',1,inplace=True)
champ_df
```

```python
tmp=pd.DataFrame()
for i in range(len(champ_df['stats'])):
    a=dict(champ_df['stats'][i])
    b=pd.DataFrame(a,index=[i])
    tmp=pd.concat([tmp,b])
tmp
```

```python
champ_df=pd.DataFrame({
    'version':champ_df['version'],
    'name':champ_df['id'],
    'champId':champ_df['key']
})
champ_df
```

```python
champ=pd.concat([champ_df,tmp],1)
champ
```

```python
champ.to_csv('D:/Final_Project/data/champion_20.csv',encoding='utf-8-sig')
```
