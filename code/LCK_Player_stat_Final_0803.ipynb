# Bang

```python
# 접속하기
player = 'Bang'
url = 'https://oracleselixir.com/player/'+ player + '/statsByYear'
wd.get(url)
```

```python
# 스탯부분 크롤링
Bang = wd.find_elements_by_css_selector('#root > div > div.sc-kRiODY.eEHzwh.mainBody > div.sc-cCgWSc.eCsPXj.mainContent > div.sc-ezrmkQ.bbVJqh > div.sc-iihChq.ciRLBw')[0].text
Bang = Bang.split('\n')
Bang
```

```python
# 리스트 나누기
def list_div(list, n):
    return [list[i:i+n] for i in range(0, len(list), n)]

Bang_div = list_div(Bang, 16) # 16 = len(col)
Bang_div
```

```python
# columns 부분 크롤링
col = wd.find_elements_by_css_selector('#root > div > div.sc-kRiODY.eEHzwh.mainBody > div.sc-cCgWSc.eCsPXj.mainContent > div.sc-ezrmkQ.bbVJqh > div.sc-fvxoyt.hztGkq')[0].text
col = col.split('\n')
col
```

```python
# 데이터프레임 생성
Bang_df = pd.DataFrame(list(Bang_div), columns=col)
Bang_df
```

```python
# 나이 컬럼 추가(만 나이(생일무관) : ex) 1996년생은 2021년에 25살)
Bang_df['Year'] = pd.to_numeric(Bang_df['Year'])
Bang_df['age'] = (Bang_df['Year'] - 1996)
Bang_df
```

```python
# 숫자형으로 변환
Bang_df['W%'] = pd.to_numeric(Bang_df['W%'].str.replace('%', '').replace('-', np.nan))
Bang_df['DMG%'] = pd.to_numeric(Bang_df['DMG%'].str.replace('%', '').replace('-', np.nan))
Bang_df['DTH%'] = pd.to_numeric(Bang_df['DTH%'].str.replace('%', '').replace('-', np.nan))
Bang_df['GOLD%'] = pd.to_numeric(Bang_df['GOLD%'].str.replace('%', '').replace('-', np.nan))
Bang_df['KP'] = pd.to_numeric(Bang_df['KP'].str.replace('%', '').replace('-', np.nan))
Bang_df.rename(columns={"KP":"KP%"}, inplace = True)
Bang_df.replace('-', np.nan, inplace = True)
Bang_df = Bang_df.astype({'GP':np.float,
               'KDA':np.float,
                'GD10':np.float,
                'XPD10':np.float,
                'CSPM':np.float,
                'DPM':np.float,
                'WPM':np.float,
                'WCPM':np.float})
Bang_df
```

```python
Bang_df.info()
```

```python
# 엑셀로 저장
Bang_df.to_excel('../data/Bang_data.xlsx', index = False)
```

```python
# 저장된 데이터 확인
Bang_df1 = pd.read_excel('../data/Bang_data.xlsx')
Bang_df1
```
