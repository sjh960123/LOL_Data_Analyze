```python
import pandas as pd
import time
import json
import requests
from tqdm import tqdm_notebook
```

```python
api_key='변경함'
api_url_challenger = 'https://kr.api.riotgames.com/lol/league/v4/challengerleagues/by-queue/RANKED_SOLO_5x5?api_key=' + api_key
# 챌린저 데이터 가져와서 데이터프레임으로 변환
r = requests.get(api_url_challenger)
r.json()
```

```python
league_df = pd.DataFrame(r.json())

# entries 데이터프레임으로 변환하여 추가
league_df.reset_index(inplace=True)
league_entries_df = pd.DataFrame(dict(league_df['entries'])).T
league_df = pd.concat([league_df, league_entries_df], axis=1)

league_df = league_df.drop(['index', 'rank'], axis=1)
league_df.info()
```

```python
league_df
```

```python
league_df.sort_values('leaguePoints', ascending=False, inplace=True)
league_df.reset_index(inplace=True)
league_df.drop('index',1,inplace=True)
league_df.to_csv('D:/Final_Project/data/challenger_list.csv',encoding='utf-8-sig')
```

```python
df=pd.read_csv('D:/Final_Project/data/challenger_list.csv',index_col=0)
df
```

```python
a=[]
for i in tqdm_notebook(range(0,300)):
    api_account_url='https://kr.api.riotgames.com/lol/summoner/v4/summoners/by-name/'+df['summonerName'][i]+'?api_key='+api_key
    r=requests.get(api_account_url)
    while r.status_code!=200: # 요청 제한 또는 오류로 인해 정상적으로 받아오지 않는 상태라면, 5초 간 시간을 지연
        time.sleep(5)
        r = requests.get(api_account_url)
    account_id=r.json()['accountId']
    a.append(account_id)
df['account_id']=a
df
```

```python
season = str(13)

match_info_df = pd.DataFrame()
for account_id in tqdm_notebook(df['account_id']):
    api_url = 'https://kr.api.riotgames.com/lol/match/v4/matchlists/by-account/' + account_id + \
                  '?season=' + season + '&api_key=' + api_key
    r = requests.get(api_url)
    while r.status_code!=200: # 요청 제한 또는 오류로 인해 정상적으로 받아오지 않는 상태라면, 5초 간 시간을 지연
        time.sleep(10)
        r = requests.get(api_url)
        
    match_info_df = pd.concat([match_info_df, pd.DataFrame(r.json()['matches'])])
match_info_df.to_csv('d:/Final_Project/data/match_id.csv',encoding='utf-8-sig') 
```
