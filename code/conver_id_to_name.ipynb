```python
import pandas as pd

match_20=pd.read_csv('D:/Final_Project/data/match_20_data.csv',index_col=0)
match_21=pd.read_csv('D:/Final_Project/data/match_21_data.csv',index_col=0)
champ_20=pd.read_csv('D:/Final_Project/data/champion_20.csv',index_col=0)
champ_21=pd.read_csv('D:/Final_Project/data/champion_21.csv',index_col=0)
spell=pd.read_csv('D:/Final_Project/data/spellid.csv',index_col=0)
```

```python
def champId20_to_name(id):
    champ_20_tmp=champ_20.set_index('champId')
    tmp=dict(champ_20_tmp['name'])
    switcher=tmp
    return switcher.get(id)

def champId21_to_name(id):
    champ_21_tmp=champ_21.set_index('champId')
    tmp=dict(champ_21_tmp['name'])
    switcher=tmp
    return switcher.get(id)
def spellId_to_name(id):
    spell_tmp=spell.set_index('spellId')
    tmp=dict(spell_tmp['name'])
    switcher=tmp
    return switcher.get(id)
```

```python
from tqdm import tqdm_notebook
```

```python
for x in (2,12,22,32,42,52,62,72,82,92):
    for i in tqdm_notebook(range(len(match_20))):
        match_20.iloc[i,x]=champId20_to_name(match_20.iloc[i,x])
match_20
```

```python
for x in (2,12,22,32,42,52,62,72,82,92):
    for i in tqdm_notebook(range(len(match_21))):
        match_21.iloc[i,x]=champId21_to_name(match_21.iloc[i,x])
match_21
```

```python
for x in (3,4,13,14,23,24,33,34,43,44,53,54,63,64,73,74,83,84,93,94):
    for i in tqdm_notebook(range(len(match_20))):
        match_20.iloc[i,x]=spellId_to_name(match_20.iloc[i,x])
match_20    
```

```python
for x in (3,4,13,14,23,24,33,34,43,44,53,54,63,64,73,74,83,84,93,94):
    for i in tqdm_notebook(range(len(match_21))):
        match_21.iloc[i,x]=spellId_to_name(match_21.iloc[i,x])
match_21    
```
